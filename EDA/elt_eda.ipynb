```python
df_t = df.copy()
```

```python
df_t['date'] = df_t['time'].apply(lambda x: datetime.datetime.utcfromtimestamp(x / 1000).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
df_t.head()
```

```python
df_t.shape
```

```python

df_t['date'] = pd.to_datetime(df_t['date'], errors='coerce')
df_t['year'] = df_t['date'].dt.year
```

```python
df_t.head()
```

```python
df_filtrado = df_t[df_t['gmap_id'] == "0x89e40887d6d58239:0xa7c8ea10076dba64"]
dimensiones = df_filtrado.shape
```

```python
df_filtrado
```

```python
dimensiones
```

```python
df_avg_rating = df_filtrado.groupby('year')['rating'].mean()

# Graficar
plt.figure(figsize=(10, 5))
plt.plot(df_avg_rating.index, df_avg_rating.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Rating per Year')
plt.grid(True)
plt.show()
```

```python
resumen = df_filtrado.groupby('year').agg(
    avg_rating=('rating', 'mean'),   # Promedio de rating
    total_votes=('year', 'sum') # Total de votos
).reset_index()
```
